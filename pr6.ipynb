Мы импортировали с помощью функции read_csv и получился датасет по наименованию переменной

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas  as pd
spotify_songs2 = pd.read_csv("/Users/veseliy.hlebushek/Desktop/spotify_songs2.csv")
spotify_songs2
```

На линейном графике отображена статистика популярности последних пяти треков из набора данных "spotify_songs2". Горизонтальная ось представляет собой названия треков, а вертикальная ось - уровень их популярности. График был построен с использованием библиотеки matplotlib в Python.
Из графика видно, как меняется популярность каждого трека. Анализируя последние пять треков, можно сделать вывод о том, что на данный момент наиболее популярен первый трек или менее популярен третий трек среди пользователей Spotify.

```python
plt.figure(figsize=(20,10))
plt.plot(spotify_songs2['track_name'].tail(5),spotify_songs2['track_popularity'].tail(5), label = 'Популярность трека')
plt.title('Статистика популярности треков')
plt.xlabel('Названиие трека')
plt.ylabel('Популярность')
plt.legend()
plt.show()
```

На линейном  графике представлена статистика разговорчивости (speechiness) последних пяти треков из набора данных "spotify_songs2". Горизонтальная ось отображает названия треков, а вертикальная ось представляет уровень разговорчивости каждого трека. График построен с использованием библиотеки matplotlib в Python, используя фиолетовый цвет для линии и маркеры для выделения точек данных.
Из графика можно сделать выводы о различной степени разговорчивости последних пяти треков. Анализируя эту метрику, можно выделить треки с более высоким уровнем (4 трек)разговорчивости и те, где она ниже (5 трек). Это может быть полезным для оценки стиля или характера каждого трека в контексте его разговорчивости.

```python
plt.figure(figsize=(20,10))
plt.plot(spotify_songs2['track_name'].tail(5),spotify_songs2['speechiness'].tail(5), label = 'Разговорчивость', color = "purple", marker='o', markersize=7)
plt.title('Статистика разговорчивости треков')
plt.xlabel('Названиие трека')
plt.ylabel('Разговорчивость')
plt.legend()
plt.grid()
plt.show()
```

На данном графике представлена статистика популярности последних пяти треков из набора данных "spotify_songs2". Линейный график (линия фиолетового цвета) отображает уровень популярности каждого трека по горизонтальной оси (названия треков) и вертикальной оси (уровень популярности). Дополнительно, с использованием столбчатой диаграммы (столбцы розового цвета), выделены значения популярности треков.
Из графика можно сделать выводы о сравнительной популярности этих треков. Линейный график демонстрирует изменения уровня популярности, а столбчатая диаграмма выделяет конкретные значения для лучшего восприятия различий между треками. Это позволяет легче сравнивать популярность треков и выделять те, которые наиболее выделяются среди последних пяти.

```python
plt.figure(figsize=(20,10))
plt.plot(spotify_songs2['track_name'].tail(5),spotify_songs2['track_popularity'].tail(5), label = 'Популярность трека', color = "purple")
plt.bar(spotify_songs2['track_name'].tail(5),spotify_songs2['track_popularity'].tail(5), label = 'Популярность трека столбцы', color = 'pink')
plt.title('Статистика популярности треков')
plt.xlabel('Названиие трека')
plt.ylabel('Популярность')
plt.legend()
plt.show()
```

На столбчатой диаграмме представлена статистика популярности последних пяти треков из набора данных "spotify_songs2". Использованы столбчатая диаграмма (красные столбцы) и легенда для обозначения уровня популярности каждого трека.
Из графика видно, как изменяется уровень популярности каждого трека. Красные столбцы предоставляют визуальное представление о популярности каждого трека в сравнении друг с другом. Легенда добавляет пояснения к графику.
Визуальный анализ этого графика может помочь выделить треки с наивысшей (1 трек) и наименьшей (3 трек) популярностью в последних пяти треках, что может быть полезно для понимания динамики изменения популярности в этом конкретном временном интервале.

```python
plt.figure(figsize=(20,10))
plt.bar(spotify_songs2['track_name'].tail(5),spotify_songs2['track_popularity'].tail(5), label = 'Популярность трека', color = 'red')
plt.title('Статистика популярности треков')
plt.xlabel('Названиие трека')
plt.ylabel('Популярность')
plt.legend()
plt.show()
```

На данном графике представлена статистика популярности и исполнителей последних пяти треков из набора данных "spotify_songs2". Линейный график (фиолетовая линия с маркерами) отображает уровень популярности каждого трека, используя левую вертикальную ось. Столбчатая диаграмма (розовые столбцы) представляет исполнителей для каждого трека, используя правую вертикальную ось.
Из графика можно сделать выводы о соотношении популярности треков и их исполнителей. Визуально можно оценить, есть ли какие-то общие тенденции между популярностью треков и исполнителями в последних пяти записях

```python
plt.figure(figsize=(20,10))
plt.plot(spotify_songs2['track_name'].tail(5),spotify_songs2['track_popularity'].tail(5), label = 'Популярность трека', color = "purple", markersize=7, marker='o')
plt.bar(spotify_songs2['track_name'].tail(5),spotify_songs2['track_artist'].tail(5), label = 'Исполнитель', color = 'pink')
plt.title('Статистика популярности треков')
plt.xlabel('Названиие трека')
plt.ylabel('Популярность')
plt.legend()
plt.show()
```

На данном графике представлена статистика популярности и исполнителей последних пяти треков из набора данных "spotify_songs2". Использованы две столбчатые диаграммы: красные столбцы обозначают уровень популярности каждого трека, а фиолетовые столбцы представляют исполнителей для каждого трека.
Из графика видно, как изменяется уровень популярности каждого трека и какому исполнителю принадлежит каждый из этих треков.

```python
plt.figure(figsize=(20,10))
plt.bar(spotify_songs2['track_name'].tail(5),spotify_songs2['track_popularity'].tail(5), label = 'Популярность трека', color = 'red')
plt.bar(spotify_songs2['track_name'].tail(5),spotify_songs2['track_artist'].tail(5), label = 'Исполнитель', color = 'purple')
plt.legend()
plt.xlabel('Названиие трека')
plt.ylabel('Исполнитель')
plt.show()
```

На данной круговой диаграмме представлено соотношение различных режимов в данных набора "spotify_songs2". Каждый сегмент круга представляет собой определенный режим, а процентные значения на графике указывают долю каждого режима в общем объеме.
Из анализа графика можно сделать выводы о том, какие режимы преобладают в данных и какова их относительная частота. Это может быть полезной информацией для понимания характеристик музыкальных треков в выборке.

```python
mode = spotify_songs2['mode'].value_counts()
plt.pie(mode, labels=mode.index, autopct='%1.1f%%', colors = ['orange','pink'])
plt.title('Соотношение режима')
plt.show()
```

На представленном круговом графике отражено соотношение различных ключей в данных набора "spotify_songs2". Каждый сегмент круга представляет собой конкретный музыкальный ключ, а процентные значения указывают на их относительную частоту в общем объеме.
Из анализа графика можно сделать выводы о том, какие музыкальные ключи преобладают в данных и какова их относительная частота. Это может быть полезной информацией для понимания предпочтений в музыкальных ключах в выборке.

```python
key = spotify_songs2['key'].value_counts()
valid_keys = key[key.index <= 5]
labels = valid_keys.index
plt.pie(valid_keys, labels=labels, autopct='%1.1f%%', colors = ['orange','pink','red','purple','blue','green'])
plt.title('Ключ')
plt.show()
```

На данной диаграмме рассеяния представлено распределение популярности треков в зависимости от исполнителя. Каждая точка на графике представляет собой сочетание уровня популярности и конкретного исполнителя для первых пяти треков из набора данных "spotify_songs2".
Из анализа графика можно сделать выводы о том, как связаны популярность трека и исполнитель в выборке. 

```python
plt.figure(figsize=(7,4))
plt.scatter(spotify_songs2['track_popularity'].head(5), spotify_songs2['track_artist'].head(5), color = 'purple')
plt.title('Связь между исполнителем и популярностью')
plt.xlabel('Популярность')
plt.ylabel('Исполнитель')
plt.show()
```

На данной диаграмме рассеяния представлено распределение популярности треков в зависимости от названия альбома для первых пяти треков из набора данных "spotify_songs2". Каждая точка на графике представляет собой сочетание уровня популярности и названия альбома.
